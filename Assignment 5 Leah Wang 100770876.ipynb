```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
def visualize(filename):
    data=pd.read_csv(filepath_or_buffer = filename)
    fig = plt.figure(figsize=(15,10)); ax = fig.add_subplot(1,1,1)
    ax.plot(data.mean(),'g',label='one')
    ax.plot(data.min(),'r',label='two')
    ax.plot(data.max(),'y',label='three')
    plt.legend(['Avg','Min','Max'])
    ax.set_xticklabels(i for i in range(40))
    plt.ylabel('Value')
    plt.xlabel('Feature#')
    plt.show()  
```

```python
# find all the filenames 
```

```python
import glob
dataset=sorted(glob.glob('data*.csv'))
print(dataset)
    
```

```python
# use for loop to iterate over a sequence
# call the function(visulize)
```

```python
for datasets in dataset[:4]:
    print("/n")
    print(datasets)
    visualize(datasets)

```

```python
def identify_issues(filename):
    data = pd.read_csv(filepath_or_buffer = filename)
    if max(data.iloc[:,0])==0 and max(data.iloc[:,20])==20:
        print("I doubt the integrity of this dataset")
    elif sum(data.min())==0:
        print("Why is this happening?")
    else:
        print("Dataset looks good")
```

```python
for datasets in dataset[:4]:
    print("/n")
    print(datasets)
    identify_issues(datasets)
```
